```python
def bessell (cx,I0):
    aa = 2*pi*a/(lamb*sqrt(a**2+LZ**2))
    arg = np.multiply(aa,cx)
    bessell=np.zeros(len(cx))
    for i in range(len(arg)):
        bessell[i] = jv(0,arg[i]**2)
    return(np.multiply(I0,bessell))
```

```python
plt.plot(cx,np.abs(am1)[center],'d-',label = "Numerical")
plt.plot(cx,np.abs(bessell(kx,1)),'d-',label = "Analytic")
plt.legend(frameon = True)
plt.title("Abs(E) for $z = 100 mm $ and $ a = 2 mm $",fontsize = 15)
plt.xlim(-.001,.001)
#plt.ylim(.96,1.01)
plt.show()
```
